```python
x=np.linspace(1,14,100)
x
```

```python
y1 = np.sin(x)
y2 = 2*np.sin(x)
y3 = 3*np.sin(x)
y4 = 4*np.sin(x)

plt.figure(figsize=(10,6))
plt.plot(x,y1,x,y2,x,y3,x,y4,)
plt.show()
```

```python
# seaborn 라이브러리 임포트
import seaborn as sns

# 그래프 전반의 스타일을 'dark' 테마로 설정
# 다른 스타일 옵션: 'darkgrid', 'whitegrid', 'white', 'ticks'
sns.set_style('darkgrid')

# x와 y1–y4 데이터를 한 번의 호출로 선 그래프 그리기
# 인자 순서: (x, y1), (x, y2), (x, y3), (x, y4)
# 기본 색상·스타일·마커가 적용되며, 필요 시 개별 스타일 지정 가능
plt.plot(
    x, y1,  # 첫 번째 선: x vs y1
    x, y2,  # 두 번째 선: x vs y2
    x, y3,  # 세 번째 선: x vs y3
    x, y4   # 네 번째 선: x vs y4
)

# 보통 여러 선을 그릴 땐 레전드를 추가하여 구분하는 것이 좋습니다.
# 예) plt.legend(['y1', 'y2', 'y3', 'y4'])
```

```python
sns.set_style('whitegrid')
plt.plot(x,y1,x,y2,x,y3,x,y4,)
sns.despine(top=True, left=True)
```

```python
# seaborn 그래프 컨텍스트를 'talk'로 설정합니다.
# 가능한 옵션:
#  • 'paper'   – 논문용 작은 폰트·두께
#  • 'notebook'– 노트북 환경 기본 설정
#  • 'talk'    – 프레젠테이션·토크용으로 폰트·선 두께를 키움
#  • 'poster'  – 포스터용으로 훨씬 크게 설정
sns.set_context('talk')

# x와 y1~y4 데이터를 한 번에 선 그래프로 그립니다.
# 각각 (x, y1), (x, y2), (x, y3), (x, y4) 순서로 플롯됨
plt.plot(
    x, y1,  # 첫 번째 선: x vs y1
    x, y2,  # 두 번째 선: x vs y2
    x, y3,  # 세 번째 선: x vs y3
    x, y4   # 네 번째 선: x vs y4
)

# (추가 권장) 여러 선을 구분하기 위해 범례를 추가할 수 있습니다.
# 예) plt.legend(['Series 1', 'Series 2', 'Series 3', 'Series 4'])
```

```python
sns.set_context('talk', font_scale=0.5)
plt.plot(x,y1,x,y2,x,y3,x,y4,)
```

```python
tips =sns.load_dataset('tips')
tips.head(5)
```

```python
sns.histplot(x=tips['total_bill'])
```

```python
# seaborn의 kdeplot 함수를 사용해 커널 밀도 추정(KDE) 플롯을 그립니다.
# x 인자에는 분포를 보고 싶은 데이터 시리즈를 지정합니다.
# 여기서는 tips 데이터프레임의 'total_bill' 열(식사 총액)을 부드러운 곡선 형태로 시각화합니다.
sns.kdeplot(x=tips['total_bill'])
```

```python
sns.boxplot(x=tips['total_bill'])
```

```python
sns.boxplot(x=tips['day'], y=tips['total_bill'])
sns.despine(offset=10)
```

```python
sns.boxplot(x=tips['day'], y=tips['total_bill'], palette='Set2')
sns.despine(offset=10)
```

```python
sns.boxplot(data=tips, x='day', y='tip' ,palette='Set3')
```

```python
sns.boxplot(data=tips, x='day', y='tip', hue='sex' ,palette='Set3')
```

```python
# Swarm plot은 범주형 변수(category)별로 개별 데이터 포인트를 모두 표현하면서도 서로 겹치지 않도록 옆으로 약간씩 퍼뜨려(jitter) 그리는 산점도 형태의 그래프입니다.
sns.swarmplot(data=tips, x='day', y='total_bill', palette='Set2')
```

```python
sns.swarmplot(data=tips, x='day', y='total_bill', palette='Set2')
sns.boxplot(data=tips, x='day', y='total_bill', palette='husl')
```

```python
sns.set_style('darkgrid')
sns.lmplot(data=tips, x='total_bill', y='tip')
```

```python
# seaborn의 lmplot: 'total_bill'과 'tip' 간의 관계를 산점도로 그립니다.
# fit_reg=False를 지정해 회귀선(regression line)은 표시하지 않습니다.
sns.lmplot(
    data=tips,           # 사용할 데이터프레임
    x='total_bill',      # x축 변수: 식사 총액
    y='tip',             # y축 변수: 팁 금액
    fit_reg=False        # 회귀선을 그리지 않음 (True면 기본 선형 회귀선 표시)
)

# (선택) 그래프를 화면에 표시
plt.show()
```

```python
sns.lmplot(data=tips, x='total_bill', y='tip', hue='smoker')
```

```python
sns.lmplot(data=tips, x='total_bill', y='tip', hue='sex')
```

```python
sns.lmplot(data=tips, x='total_bill', y='tip', col='sex')
```
