```python
# 주어진 y값 리스트 [1, 4, 9, 16]를 인덱스(0,1,2,3)에 따라 선 그래프로 그립니다.
plt.plot([1, 4, 9, 16])
```

```python
# 0부터 6까지 연속된 정수 배열 생성 (t = [0, 1, 2, 3, 4, 5, 6])
t = np.arange(7)

# 각 t 값에 대응되는 y 값 리스트 정의
y = [1, 4, 5, 8, 9, 5, 3]

# t와 y를 함께 출력하여 값 확인
t, y
```

```python
# 가로 10인치, 세로 6인치 크기의 새로운 그래프 창(Figure)을 생성합니다.
plt.figure(figsize=(10, 6))

# t (x축)와 y (y축) 데이터를 빨간 선(color='red')으로 그립니다.
plt.plot(t, y, color='red')
```

```python
# 다양한 색상(color) 지정 방법 예시:
#  • 이름 문자열: 'red', 'green', 'blue', 'orange' 등
#  • HEX 코드: '#1f77b4', '#ff7f0e', '#2ca02c' 등
#  • RGB 튜플: (0.1, 0.2, 0.5), (1.0, 0.0, 0.5) 등
#  • 그레이스케일 문자열: '0.0'~'1.0' (검정→흰색)

# 다양한 선 스타일(linestyle) 예시:
#  • 실선: '-' 또는 'solid'
#  • 파선: '--' 또는 'dashed'
#  • 점선: ':' 또는 'dotted'
#  • 대시-닷: '-.' 또는 'dashdot'

# 예1: 녹색 점선 (이전에 보신 예)
plt.plot(t, y, color='green', linestyle='dotted')
```

```python
# t (x축)와 y (y축) 데이터를 녹색 점선(dotted)으로 그리고,
# 각 데이터 포인트 위치에 원형 마커(marker='o')를 표시합니다.
plt.plot(t, y,
         color='green',     # 선 색상: 이름 문자열
         linestyle='dotted',# 점선 스타일
         marker='o')        # 마커: 'o'=원형

# 추가 tip – marker에 쓸 수 있는 몇 가지 예시:
#  • 's' : 정사각형(square)
#  • '^' : 위쪽 삼각형(triangle_up)
#  • '*' : 별 모양(star)
#  • 'D' : 다이아몬드(diamond)
#  • '.' : 점(point)
```

```python
# t (x축)와 y (y축) 데이터를 녹색 점선으로 그리고,
# 각 데이터 포인트 위치에 원형 마커(marker='o')를 표시합니다.
# 마커 내부 색상(markerfacecolor)을 빨간색으로, 마커 크기(markersize)를 12로 설정합니다.
plt.plot(t, y,
         color='green',         # 선 색상: 'green' (이름 문자열)
         linestyle='dotted',    # 선 스타일: 'dotted' (점선)
         marker='o',            # 마커 모양: 'o' (원형)
         markerfacecolor='red', # 마커 내부 색상: 'red' (이름 문자열)
         markersize=12)         # 마커 크기: 12 (포인트 단위)
```

```python
# x축, y축 데이터 정의
x = [10, 20, 30, 40]
y = [1,  4,  9, 16]

# 선 그래프 그리기
plt.plot(
    x, y,
    c='g',    # 선 색상: 'g'는 green의 축약형
    lw=5,     # 선 두께(linewidth): 5포인트
    marker='o',  # 데이터 포인트마다 원형 마커 표시
    ms=15,    # 마커 크기(markersize)의 축약형: 15포인트
    mfc='r'   # 마커 내부 색상(markerfacecolor): 'r'은 red의 축약형
)

# 그래프 제목 설정
plt.title("<예시>")  # 제목 텍스트는 '<예시>'
```

```python
# x, y 데이터에 대해 선 그래프를 그립니다.
# c='g'는 green의 축약형 색상, lw=5는 선 두께, marker='o'는 원형 마커,
# ms=15는 마커 크기, mfc='r'는 마커 내부 색상(red의 축약형)
plt.plot(x, y, c='g', lw=5, marker='o', ms=15, mfc='r')

# 그래프의 제목을 '<예시>'로 설정합니다.
plt.title("<예시>")

# x축의 표시 범위를 0에서 50까지로 제한합니다.
plt.xlim(0, 50)

# y축의 표시 범위를 -10에서 30까지로 제한합니다.
plt.ylim(-10, 30)
```

```python
# 0부터 5 미만까지 0.2 간격으로 연속된 값 배열 생성 (t = [0. , 0.2, ..., 4.8])
t = np.arange(0, 5, 0.2)

# 가로 10인치, 세로 6인치 크기의 새 Figure 생성
plt.figure(figsize=(10, 6))

# 그래프 제목 설정
plt.title('여러개의 선 그리기')

# 1. t vs t: 빨간색 점선(r--)으로 직선 그리기
#    'r' = red, '--' = dashed line
plt.plot(t, t, 'r--')

# 2. t vs t**2: 파란색 사각형 마커와 점선(bs:)으로 2차 곡선 그리기
#    'b' = blue, 's' = square marker, ':' = dotted line
plt.plot(t, t**2, 'bs:')

# 3. t vs t**3: 초록색 위쪽 삼각형 마커와 실선(g^-)으로 3차 곡선 그리기
#    'g' = green, '^' = triangle_up marker, '-' = solid line
plt.plot(t, t**3, 'g^-')
```

```python
# 한 번의 plt.plot 호출로 여러 곡선을 그리는 예시
plt.plot(
    t, t,    'r--',  # t vs t: 빨간색(red) 파선(dashed line)
    t, t**2, 'bs:',  # t vs t^2: 파란색(blue) 정사각형(square) 마커 + 점선(dotted line)
    t, t**3, 'g^-'   # t vs t^3: 초록색(green) 위쪽 삼각형(triangle_up) 마커 + 실선(solid line)
)
```

```python
np.random.seed(0)
people=['지호','준호','재홍','상원']
y = np.arange(len(people))
performance = 3+ 10 * np.random.rand(len(people))
performance
```

```python
# 그래프 제목을 'Bar Chart'로 설정
plt.title('Bar Chart')

# 가로 방향(horizontal) 막대 그래프 그리기
# 첫번째 인자 y: 막대의 y축 위치 리스트
# 두번째 인자 performance: 각 막대의 길이(성능) 값 리스트
# alpha=0.4: 막대의 투명도 설정 (0.0~1.0)
plt.barh(y, performance, alpha=0.4)

# y축 눈금 위치와 레이블을 y 리스트와 people 리스트로 설정
# y 위치마다 대응되는 사람 이름(people)이 표시됨
plt.yticks(y, people)

# x축 레이블을 'xlabel'로 설정
plt.xlabel('xlabel')

# y축 레이블을 'ylabel'로 설정
plt.ylabel('ylabel')
```
